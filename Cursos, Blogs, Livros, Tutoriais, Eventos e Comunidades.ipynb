[Automated Trading with R: Quantitative Research and Platform Development](http://gen.lib.rus.ec/search.php?req=Automated-Trading-with-R-Quantitative-Research-and-Platform-Development&lg_topic=libgen&open=0&view=simple&res=25&phrase=1&column=def); livro bem técnico sobre negociação automatizada em tradings no mercado de ações, começando com um pouco de matemática e passando para a computação e execução. Quem não gosta de ganhar dinheiro? No mercado de renda variável as coisas são um pouco complicadas mas com dedicação é possível aprender algumas estratégias e começar a ter retorno financeiro ao mês com um pequeno capital investido. Este livro oferece uma compreensão inicial dos mecanismos internos de um sistema de negociação automatizado com a linguagem R, permite que o trader acesse suas bibliotecas de pacotes para otimizar estratégias, gerando decisões de negociação, simulando o desempenho estratégico e até mesmo critérios de aprendizado de máquina para validar estatística no contexto de séries temporais.

[Como mentir com estatstica](http://professoresdematematica.com.br/wa_files/2_20LIVRO_20COMO_20MENTIR_20COM_20ESTAT_C3_8DSTICAS.pdf); como mentir com estatística é considerado um clássico que foi primeiramente publicado nos Estados Unidos em 1954. Desde então teve várias publicações no exterior e no Brasil, eu li ele todo e trata do assunto com muito bom humor e as ilustrações são bem hilárias, os exemplos costumam ser da década de 90 mas os conceitos ainda são extremamente válidos e devem servir como referência para todos que realizam análises de dados. O autor consegue mostrar de forma muito direta como números que parecem tão fortes são na realidade frágeis, quando não totalmente falsos, este é um livro de cabeceira para dar uma folheada antes de dormir.

[O Andar do Bêbado - Como o Acaso Determina Nossas Vidas](http://amicaru.com.br/ebooks/ebook168.pdf); um colega de grupo de estudos me indicou este livro um tempo atrás e confesso que preciso terminar de ler. O livro aborda a aleatoriedade em nossas vidas ao ponto de não percebermos o quanto o acaso nos interfere o tempo todo. O autor, Leonard Mlodinow, cita exemplos e pesquisas presentes no mercado financeiro, nos esportes e até em Hollywood. O livro traz elucidações de como conviver melhor com fatores que não podemos controlar, "muitas coisas são tão previsíveis quanto o próximo passo de um bêbado depois de uma noitada...", cita o autor. 

Para finalizar este tópico, um dos vários casos descritos no livro é o do problema do iPod Shuffle, que diz que muitos usuários deste iPod duvidavam da aleatoriedade com que as músicas eram tocadas, pois algumas vezes um mesmo artista ou música tocava mais de uma vez. A saída encontrada por Steve Jobs, segundo o livro, foi reprogramar o iPod para que evitasse repetições e deixá-lo menos aleatório para parecer mais aleatório.
