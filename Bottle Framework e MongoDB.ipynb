# Manipulando a base de dados no Python

A mesma manipulação da base de dados realizada no prompt é possível diretamente pelo Python (ou utilizando o [Jupyter Notebook](https://pablojuancruz.wordpress.com/2014/09/03/configurando-ambiente-mongodb-no-windows/)) sem a necessidade da conexão ao servidor local pelo prompt de comando digitando mongod para a execução do serviço.

O arquivo amostra.json contém os dois registros, Rafael e nathan, como abaixo. Criaremos basicamente o mesmo arquivo e a única diferença seriam os id`s, pois eles são inseridos automaticamente a cada novo documento criado.

Importaremos pprint e pymongo, criaremos conexão ao servidor local (porta 27017) e automaticamente serão criadas (se não existirem ainda) a base de dados (names) e a coleção (mynames).

```python
%matplotlib inline
from pprint import pprint
from pymongo import MongoClient

client = MongoClient('localhost:27017') ## 27017 é a porta padrão para a instalação do Mongo DB
db = client.names
collection = db.mynames
print(collection)
```

Passamos o nome da base de dados em forma de dicionário.

```python
db = client['names']
db
```

Verificamos quais base de dados já foram criadas antes e estão disponíveis.

```python
client.list_database_names()
```

Se precisar saber quais coleções já existem, fazemos da seguinte forma.

```python
db.list_collection_names()
```

Cada registro em uma tabela tem os mesmos campos (geralmente com valores diferentes) na mesma ordem, enquanto cada documento em uma coleção pode ter campos completamente diferentes dos outros documentos.

Os documentos são dicionários Python que podem ter strings como chaves e podem conter vários tipos primitivos tais como inteiros (int), pontos flutuantes (float), unicode, datetime, bem como outros documentos e também arrays (listas do Python).

Para inserir os dados que precisamos no MongoDB é necessário criar um dicionário e chamar a função .insert() no objeto de coleção, como abaixo.

```python
doc = [{"name":"Rafael","email":"rafael@gmail.com"}, {"name":"nathan","email":"nathan@10gen.com"}]

collection.insert(doc)
```

Com os dois registros criados o sistema inclui automaticamente um número de objeto aleatório de identificação (_id_) para cada um deles. Apresento pelos menos três maneiras de visualizar os documentos criados, podemos simplesmente digitar doc...

```python
doc
```

... outra forma de consultar e obter a saída um pouco diferente seria por esta função abaixo. Seria mais interessante usá-la se a quantidade de registros for grande, pois cada registro é exibido linha a linha, facilitando a visualização (caso não contenha muitos campos).

E como terceira e última forma, podemos usar a função lista delimitando o número de outputs se a mesma for muito longa.

```python
list(collection.find()) # [:2]: para delimitar os registros.
```

O método find_one() seleciona um único documento de uma coleção e retorna esse documento. É útil quando se está interessado apenas na primeira correspondência

A função .estimated_document_count() retorna a quantia de coleções que temos.

```python
collection.estimated_document_count()
```

Se for necessário remover um documento da coleção, devemos especificá-lo como abaixo.

Para remover toda a base de dados, poderia ser feito desta forma.

O método update_one() é usado para modificar um documento já existente. Um documento composto é passado como argumento para update_one(), a primeira parte é usada para combinar os documentos cujo qual a mudança será aplicada e na segunda parte é que contém a mudança de fato, acabei incluindo mais uma variável, a cidade de Atibaia no registro Rafael, eu podería ter criado diversas outras seguindo sempre a ordem das chaves, mas como não sou nenhum expert decidi facilitar.

```python
# Use replace_one, update_one ou update_many
collection.update_one({'name': 'Rafael'}, {"$set": {"city": "Atibaia"}})
```

```python
list(collection.find())
```

Podemos ver nossa coleção em um DataFrame de pandas da forma abaixo. DataFrame é simplesmente um conjunto de Séries (Série nada mais é que um array de 1 dimensão). Trata-se de uma estrutura de dados de 2 dimensões com colunas e linhas (uma tabela).

```python
import pandas as pd

df = pd.DataFrame(list(collection.find()))
print ('# of columns:', len(df.columns))
print ('# of observations:', len(df))
df.head()
```

Para que possamos salvar a coleção em json é necessário codificar toda a string, retornando os documentos para a codificação padrão ('utf-8'), como podemos ver logo a seguir.

```python
for column in df:
    df[column] = df[column].str.encode('utf-8')
```

Feito isso, podemos salvar os documentos no formato .jon ou também no formato .csv, visto abaixo.

```python
df.to_json("amostra.json")
```

```python
df.reset_index().to_csv("amostra.csv")
```
